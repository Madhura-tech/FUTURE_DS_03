```python
import pandas as pd

df = pd.read_csv('student_feedback.csv')
df.head()
```

```python
# Remove unwanted index column
df.drop
```

```python
# Remove unwanted index column
df.drop(columns=['Unnamed: 0'], inplace=True)

df.head()
```

```python
df.describe()
```

```python
avg_scores = df.mean().sort_values(ascending=False)
avg_scores
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(
    x=avg_scores.index,
    y=avg_scores.values
)
plt.xticks(rotation=45, ha='right')
plt.title('Average Student Feedback Scores by Criteria')
plt.ylabel('Average Rating')
plt.xlabel('Feedback Criteria')
plt.show()
```

```python
df['Overall_Score'] = df.mean(axis=1)
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Overall_Score'], bins=10, kde=True)
plt.title('Overall Student Satisfaction Distribution')
plt.xlabel('Overall Score')
plt.ylabel('Number of Students')
plt.show()
```
